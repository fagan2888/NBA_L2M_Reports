```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as sc
import matplotlib.pyplot as pl
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from cStringIO import StringIO
    %matplotlib inline
```

```python
# Function from StackOverflow: http://stackoverflow.com/a/20905381
def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = file(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos=set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password,caching=caching, check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

```python
# Locations of downloaded L2M reports.
p2015 = '/Users/JordanVani/Documents/Assorted/Data/NBA_L2M_Reports/ReportData2015_2016/'
p2016 = '/Users/JordanVani/Documents/Assorted/Data/NBA_L2M_Reports/ReportData2016_2017/'

# Get filename of all downloaded L2M Reports
f2015 = os.listdir(p2015)
f2016 = os.listdir(p2016)

# List of team names as used in the L2M reports.
t = ['Hawks', 'Celtics', 'Nets', 'Hornets', 'Bulls',
     'Cavaliers', 'Mavericks', 'Nuggets', 'Pistons', 
     'Warriors', 'Rockets', 'Pacers', 'Clippers', 
     'Lakers', 'Grizzlies', 'Heat', 'Bucks',
     'Timberwolves', 'Pelicans', 'Knicks', 'Thunder',
     'Magic', '76ers', 'Suns', 'Trail Blazers', 'Kings',
     'Spurs', 'Raptors', 'Jazz', 'Wizards']

# Create dictionaries for each year with Team: 0.
teams = dict((k, 0) for k in t)
t2015 = teams.copy()
tr2015 = teams.copy()
t2016 = teams.copy()
tr2016 = teams.copy()
```

```python
def IncorrectCalls(path, files, teams, totalreports):
    """Count all incorrect calls for each team, add total for each team.
    path - Location of L2M reports.
    files - L2M reports filename.
    teams - Dictionary to add incorrect calls to.
    tr - Dictionary to count reports for each team"""
    for f in files:
        text = convert_pdf_to_txt(path + f)
        for x in t:
            if text.count(x) > 0:
                totalreports[x] += 1
                teams[x] += text.count('\nIC')
                teams[x] += text.count('\nINC')
```

```python
# Count incorrect calls for each season to date.
IncorrectCalls(p2015, f2015, t2015, tr2015)
IncorrectCalls(p2016, f2016, t2016, tr2016)
```

### 2015 - 2016 Season

```python
# Load into dataframe, and calculate zscore.
df2015 = pd.DataFrame.from_dict(t2015, orient='index')
df2015.columns = ['Incorrect_Calls']
df2015['TotalReports'] = tr2015.values()
df2015['IC_PerReport'] = (df2015['Incorrect_Calls'] / df2015['TotalReports']).round(2)
df2015['ZScore'] = sc.mstats.zscore(df2015.IC_PerReport)
```

```python
# Plot distribution.
fig, (ax1) = pl.subplots(nrows = 1, ncols = 1, figsize = (18, 2))
ax1.axes.get_yaxis().set_ticks([])
ax1.scatter(df2015['IC_PerReport'], np.zeros(df2015['IC_PerReport'].shape), 
            label = 'NBA Team', c = 'k', s=60, alpha=0.4)
ax1.plot([df2015['IC_PerReport'].mean(), df2015['IC_PerReport'].mean()],
         [-1, 1], label = 'League Average', c = 'r', alpha=0.4)
ax1.set_xlabel('Incorrect Calls in the Last 2 Minutes', fontsize=14)
ax1.set_title('Incorrect Calls in the Last 2 Minutes for Each NBA Team (2015 - 2016 Season)',
              fontsize = 16)
ax1.legend(ncol=5, loc='upper right')
```

```python
# Save dataframe to csv.
df2015.to_csv('NBA_L2M_Report20152016.csv', encoding = 'utf-8')
print df2015.shape
df2015.sort_values('ZScore')
```
